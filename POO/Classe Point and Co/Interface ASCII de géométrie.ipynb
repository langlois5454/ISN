# Implémentation d'une interface ASCII pour un logiciel de géométrie

Vous avez travaillé jusqu'à maintenant avec les classes Point, PointNomme, Polygone, Rectangle, Carre et Triangle.

Nous vous proposons maintenant de créer un interface en mode ASCII pour manipuler ces éléments.

Vous disposez déjà de plusieurs classes prédéfinies, et accessible via des fichiers dans le cours en ligne :

+ classes Point et PointNomme : voir fichier point.py
+ classes Polygone, Rectangle, Carre, Triangle : voir fichier polygone.py
+ classe Feuille : il s'agit de la classe qui permet de "poser" sur l'espace de travail les points et divers polygones. Voir fichier feuille.py
+ classe Interface : c'est cette classe qui gère l'interaction avec l'utilisateur. Il s'agit d'une interaction utilisant des menus en mode texte. Voir fichier interface.py

Au cours des travaux à faire, vous êtes libres d'enrichir ces classes afin de répondre aux besoins de l'interface.

Dans les travaux ci-dessous, les consignes ne spécifient pas tout. c'est à dessein afin que vous soyez confrontés à des choix cornéliens (et formateurs).

Commentez et testez au fur et à mesure !!!!

# Travail 1

Consultez les fichiers, exécutez le fichier interface.py, utilisez les fonctionnalités "Créer point" et "Lister contenu feuille". Etudiez le code avec le formateur

# Travail 2

Améliorez la fonctionnalité "Créer point" de manière à ce qu'il soit impossible de créer un point de même nom qu'un point déjà créé.


# Travail 3

Implémentez la fonctionnalité "Créer polygone". Les points sont choisis par l'utilisateur via leur nom, mais c'est bien par des références aux points concernés que le construteur de Polygone est appelé. 

Nous choisirons de faire des références aux points de la feuille de travail, et non des copies.

Attention, si le polygone créé par l'utilisateur est un rectangle, ou un carre, ou encore un triangle, il faudra faire appel au constructeur correspondant. 


# Travail 4

Implémentez la fonctionnalité "Bouger point".

Attention, si un point fait partie d'un polygone, cela peut poser problème. En effet, si le polygone est un carré et que vous bougez un point, le polygone n'est plus un carré.

Vous avez alors le choix :

+ interdire de bouger un point faisdant partie d'un polygone (choix simple)
+ à la volée détecter l'incohérence, et remplacer l'instance de Carre (par exemple), par une instance de Rectangle, ou même de Polygone (choix moins simple, voire pas simple)

C'est à vous de faire ce choix. Ce choix aura des conséquences sur d'autres fonctionalités.

# Travail 9

Implémentez la fonctionnalité "Affichage", qui est particulière. En effet, l'objectif est d'afficher en mode ASCII la feuille. On veut que les points soient placés dans un espace de disons, 81 colonnes et 81 lignes. Le repère est défini de telle manière que le point (0,0) est en ligne 41 et en colonne 41. 

Vous avez le choix :

+ Soit vous pouvez calculer les paramètres d'un zoom de telle manière que le point à l'extrème droite soit affiché sur le colonne 81, le point à l'extrème gauche soit affiché sur la colonne 1, et de même pour les lignes (en recgardant les abscisses)

+ Vous considérez que, par exemple, 1 unité de texte = 1 unité de votre repère et que la fenêtre d'affichage peut bouger verticalement et horisontalement (il faut alors prévoir des options de mouvement de la fenêtre dans le menu principal)

+ vous mixez tout en permettant de définir la taille de la fenêtre d'affichage, de zoomer, et de bouger la fenêtre d'affichage

Pour cet affichage, seuls les points sont demandés. Voir ci-dessous pour afficher les polygones.
